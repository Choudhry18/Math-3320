```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta as beta_dist
```

```python
# Parameters
alpha = 20
beta = 80
num_samples = 10000
```

# 1. Find the expected conversion rate (mean of Beta)

```python
expected_conversion = alpha / (alpha + beta)
print(f"Expected conversion rate: {expected_conversion:.4f}")
```

# 2. Calculate the mode of this Beta distribution

```python
if alpha > 1 and beta > 1:
    mode = (alpha - 1) / (alpha + beta - 2)
    print(f"Mode of the Beta distribution: {mode:.4f}")
else:
    print("Mode is not defined for these parameters")
```

# 3. Simulate 10,000 random samples from this Beta distribution

```python
samples = np.random.beta(alpha, beta, num_samples)

# Estimate probability that conversion rate is greater than 25%
prob_greater_than_25 = np.mean(samples > 0.25)
print(f"Probability that conversion rate > 25%: {prob_greater_than_25:.4f}")
```

# 4. Plot histogram of sampled values

```python
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=50, alpha=0.7, density=True, color='skyblue', edgecolor='black')

# Add the theoratical Beta PDF for comparison
x = np.linspace(0, 0.5, 1000)
plt.plot(x, beta_dist.pdf(x, alpha, beta), 'r-', lw=2, label='Beta PDF')

plt.title(f'Beta Distribution with α={alpha}, β={beta}')
plt.xlabel('Conversion Rate')
plt.ylabel('Density')
plt.legend()
# The threshold line visualizes part 3
plt.axvline(x=0.25, color='green', linestyle='--', label='25% Threshold')
plt.grid(True, alpha=0.3)
plt.show()
```
