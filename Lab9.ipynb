```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Data
failure_stress = [280, 330, 335, 378, 400, 425, 450, 506, 560, 575, 580, 
                  589, 626, 637, 690, 715, 757, 800, 900]

# Q-Q plot
stats.probplot(failure_stress, dist="norm", plot=plt)
plt.title('Q-Q Plot of Fiber Failure Stress')
plt.show()
```

The points are pretty close to the linear line so we can assume normality 

```python
# One-sample t-test
t_stat, p_value = stats.ttest_1samp(failure_stress, 500)

# One-tailed test adjustment
p_value_one_tailed = p_value / 2

print(f"T-statistic: {t_stat:.4f}")
print(f"One-tailed P-value: {p_value_one_tailed:.4f}")
```

p value is greater than 0.05 for null hypthesis so it can't be rejected at this signicance level 

# Problem 2

```python
# Data
time_estimates = [69, 65, 72, 73, 59, 55, 39, 52, 67, 57, 
                  56, 50, 70, 47, 56, 45, 70, 64, 67, 53]

# Q-Q plot
stats.probplot(time_estimates, dist="norm", plot=plt)
plt.title('Q-Q Plot of Time Estimation')
plt.show()
```

The points are along the line so normality can be assumed

```python
# One-sample t-test
t_stat, p_value = stats.ttest_1samp(time_estimates, 45)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```

p is less than 0.05 so we can reject the null hypothesis at this confidence interval

# Problem 3

Null hypothesis _0: p = 0.05


Significance level:   = 0.001

=0.001

Sample proportion: 

27/267




```python
# Proportion test
p_hat = 27 / 267
p_0 = 0.05
n = 267

z_stat = (p_hat - p_0) / np.sqrt(p_0 * (1 - p_0) / n)
p_value = 1 - stats.norm.cdf(z_stat)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.6f}")
```

p-value is less than 0.0001 so we reject the null hypothesis at this confidence level

The t-test is used when dealing with means and the population standard deviation is unknown (especially for small samples).

The z-test is used for proportions because:

The sampling distribution of the sample proportion is approximately normal (central limit theorem).

The standard error for proportions is known 

No need to estimate standard deviation from the sample like we do for means.
