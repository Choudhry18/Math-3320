```python
# Re-import necessary libraries after kernel reset
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# New sample data for website time (in minutes)
web_times = np.array([12, 15, 14, 10, 13, 17, 16, 11, 15, 13, 9, 10])
n_web = len(web_times)
mean_web = np.mean(web_times)
std_web = np.std(web_times, ddof=1)

# 95% Confidence Interval
alpha_web = 0.05
t_critical_web = stats.t.ppf(1 - alpha_web/2, df=n_web - 1)
margin_of_error_web = t_critical_web * std_web / np.sqrt(n_web)
ci_web_95 = (mean_web - margin_of_error_web, mean_web + margin_of_error_web)

# Visualization
plt.figure(figsize=(8, 5))
plt.scatter(range(n_web), web_times, color='blue', label='Individual Data Points')
plt.axhline(y=mean_web, color='green', linestyle='--', label=f'Mean = {mean_web:.2f}')
plt.fill_between(range(n_web), ci_web_95[0], ci_web_95[1], color='gray', alpha=0.2, label='95% Confidence Interval')
plt.title("Time Spent on Website - 95% CI")
plt.xlabel("Sample Index")
plt.ylabel("Time (minutes)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

mean_web, std_web, 
print(f"the 95% confidence interval is {ci_web_95}")
```
