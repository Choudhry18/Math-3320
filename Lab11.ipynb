```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load data (adjust file path as needed)
df = pd.read_csv('gasticket.csv')

# Preview data
print(df.head())
```

```python
# Prepare data
X_crude = df[['crude']]  # Predictor
y_gasoline = df['gas']  # Response

# Fit regression model
model_gasoline = LinearRegression()
model_gasoline.fit(X_crude, y_gasoline)

# Predictions
y_pred_gasoline = model_gasoline.predict(X_crude)

# Scatter plot + regression line
plt.scatter(X_crude, y_gasoline, label='Data')
plt.plot(X_crude, y_pred_gasoline, color='red', label='Regression Line')
plt.xlabel('Crude Oil Price')
plt.ylabel('Gasoline Price')
plt.title('Gasoline Price vs. Crude Oil Price')
plt.legend()
plt.show()

# Coefficient of Determination (R^2)
r2_gasoline = r2_score(y_gasoline, y_pred_gasoline)
print(f"R^2 for Gasoline vs. Crude Oil: {r2_gasoline:.3f}")
```

There is a strong direct correlation so I agree with the claims

```python
# Prepare data
y_ticket = df['airfare']

# Fit regression model
model_ticket = LinearRegression()
model_ticket.fit(X_crude, y_ticket)

# Predictions
y_pred_ticket = model_ticket.predict(X_crude)

# Scatter plot + regression line
plt.scatter(X_crude, y_ticket, label='Data')
plt.plot(X_crude, y_pred_ticket, color='green', label='Regression Line')
plt.xlabel('Crude Oil Price')
plt.ylabel('Airline Ticket Price')
plt.title('Airline Ticket Price vs. Crude Oil Price')
plt.legend()
plt.show()

# Coefficient of Determination (R^2)
r2_ticket = r2_score(y_ticket, y_pred_ticket)
print(f"R^2 for Ticket vs. Crude Oil: {r2_ticket:.3f}")
```

The correlation is not that strong so I don't agree with the reporter. While this supports the idea that crude oil prices impact gasoline prices, the relationship is only moderate, suggesting that other market forces likely play a significant role.



crude oil is more strongly related to gasoline prices than airline ticket prices


Gasoline: Prices adjust more directly with crude.

Airline tickets: Influenced by fuel hedging, competition, etc
