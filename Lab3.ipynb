Q1. Simulate flipping 10 times for a 1000 tries. Done using random

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def flip10():
    headCount = 0
    for _ in range(10):
        if random.randint(0, 1) == 0:
            headCount += 1
    return headCount

head7 = 0
for _ in range(1000):
    if flip10() == 7:
        head7 += 1
print(head7/1000)
```

Q.2 using pmf to find prob of getting exactly 7 heads in 10 flips

```python
# Parameters
n = 10  # Number of coin flips
p = 0.5  # Probability of heads



# Values for k (number of heads)
k_values = np.arange(0, n + 1)

# Compute PMF
pmf_values = binom.pmf(k_values, n, p)
print(f"Probability of exactly 7 heads in 10 flips (p = 0.5):{pmf_values[7]:3f}")
# Plot PMF
plt.figure(figsize=(8, 5))
plt.vlines(k_values, 0, pmf_values, colors='b', lw=5, label='PMF')
plt.scatter(k_values, pmf_values, color='r', zorder=3)
plt.xlabel("Number of Heads (k)")
plt.ylabel("Probability")
plt.title(f"PMF of Getting k Heads in {n} Fair Coin Flips")
plt.xticks(k_values)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
n = 10
p_values = [0.3, 0.5, 0.7]
x = np.arange(0, n + 1)
num_p = len(p_values)
bar_width = 0.2

plt.figure(figsize=(8, 6))

for i, p in enumerate(p_values):
    # Offset positions for each probability to avoid overlap
    positions = x - (num_p - 1) * bar_width / 2 + i * bar_width
    pmf = binom.pmf(x, n, p)
    plt.bar(positions, pmf, width=bar_width, label=f'p = {p}', alpha=0.7)

plt.title("Binomial PMFs as Bar Graphs (n = 10)")
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.xticks(x)
plt.legend()
plt.grid(True, axis='y')
plt.show()
```
