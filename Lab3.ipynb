Q1. Simulate flipping 10 times for a 1000 tries. Done using random

```python
import random

def flip10():
    headCount = 0
    for _ in range(10):
        if random.randint(0, 1) == 0:
            headCount += 1
    return headCount

head7 = 0
for _ in range(1000):
    if flip10() == 7:
        head7 += 1
print(head7/1000)
```

Q.2 using pmf to find prob of getting exactly 7 heads in 10 flips

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10  # Number of coin flips
p = 0.5  # Probability of heads

# Values for k (number of heads)
k_values = np.arange(0, n + 1)

# Compute PMF
pmf_values = binom.pmf(k_values, n, p)

# Plot PMF
plt.figure(figsize=(8, 5))
plt.vlines(k_values, 0, pmf_values, colors='b', lw=5, label='PMF')
plt.scatter(k_values, pmf_values, color='r', zorder=3)
plt.xlabel("Number of Heads (k)")
plt.ylabel("Probability")
plt.title(f"PMF of Getting k Heads in {n} Fair Coin Flips")
plt.xticks(k_values)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
